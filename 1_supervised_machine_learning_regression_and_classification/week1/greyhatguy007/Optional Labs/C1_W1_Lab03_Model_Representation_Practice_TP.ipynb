```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use(style='./deeplearning.mplstyle')
```

```python
x_train = np.array([1.0, 2.0]) # in 1000 sqft
y_train = np.array([300, 500.0]) # in 1000s dollars
print (f"x_train = {x_train}")
print (f"y_train = {y_train}")
```

```python
# No. of training samples
print (f"x_train.shape = {x_train.shape}")
print (f"len(x_train) = {len(x_train)}")
m = x_train.shape[0]
print (f"Number of training samples = {m}")
```

```python
# i_th training sample
i = 0 
x_i, y_i = x_train[i], y_train[i]
print (f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
```

```python
plt.scatter(x_train, y_train, marker='x', c='r')
plt.title('Housing Prices')
plt.ylabel('Price (in 1000s of dollars)')
plt.ylabel('Size (in 1000 sqft)')
plt.show()
```

```python
def compute_model_output(x, w, b):
    """
    Computes the prediction of a linear model
    
    Args:
    -----
    x (ndarray (m, )): Data, m examples
    w,b (scalar)     : model parameters
    
    Returns:
    -------
    f_wb (ndarray (m,)) : target values
    """
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = x[i]*w + b 
    return f_wb
```

```python
w = 200
b = 100
print (f"w: {w}")
print (f"b: {b}")

tmp_f_wb = compute_model_output(x_train, w, b)

# plot org data points
plt.scatter(x_train, y_train, c='r', label='Actual Values')

# plot the predictions by our model
plt.plot(x_train, tmp_f_wb, c='b', label='Our Predictions')

x_i = 1.2
cost_1200sqft = w*x_i + b
print (f"Cost of 1200sqft: {cost_1200sqft:.0f} thousand dollars") 
```
